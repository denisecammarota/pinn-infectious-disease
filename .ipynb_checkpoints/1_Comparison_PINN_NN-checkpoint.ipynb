# Notebook 1: Physics Inspired vs Neural Networks

In this notebook, I will introduce the basics of physics inspired neural networks (PINNs) and neural networks for approximating the trajectory of the Susceptible-Infected (SI) model. I will compare both approaches and discuss which performs best for this task. 

## Susceptible-Infected  (SI) model

The Susceptible-Infected (SI) model is one of the simplest compartmental models in mathematical epidemiology, describing a disease for which the population is divided in only two classes: susceptible to infection and infected. In a closed poulation, the SI model is defined by the system of equations:

$$
\begin{aligned}
\frac{dS}{dt} = -\beta S I
\frac{dI}{dt} = \beta S I,
\end{aligned}
$$
where S and I are the proportions of susceptibles and infected. Considering a constant population size such that $S + I = 1$, the system above is described by only one equation for $I$:

$$
\begin{aligned}
\frac{dI}{dt} = \beta (1 - I) I.
\end{aligned}
$$

This equation, known as the logistic, has a unique analytical solution, provided that we give an initial condition. Considering that at $t = 0$, I(0) = I_0, we have that the proportion of infected over time is given by:

$$
\begin{aligned}
I(t) = \frac{I_0 e^{\beta t}}{(1 - I_0) + I_0 e^{\beta t}}
\end{aligned}
$$

While the dynamics of this system are easy and can be solved analytically without much effort, we will use it as an example to introduce and compare the performances of ordinary and physics inspired neural networks (PINNs) for predicting its dynamics.


## Ordinary Neural Networks

For predicting dynamics from ordinary neural networks, we will first import  some necessary libraries as well as the class NNetwork from the neuralnet.py file. This class implements a basic neural network composed by providing four basic parameters:
- dim_input: the dimension of our input data (number of features)
- dim_output: the dimension of our desired output (target dimension)
- dim_hidden: the number of hidden layers for our network
- dim_layer: the number of neurons in each layer (considered equal for all layers)

```python
import numpy as np
import torch
import torch.nn as nn
from torch.autograd import grad
import torch.optim as optim
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
from neuralnet import NNetwork
torch.manual_seed(1234)  # setting pytorch seed
```

We will first create some data on the SI model. First, we define a function that gives us the analytical solution of this model:

```python
@np.vectorize
def model_SI(t, beta, I0):
    # Returns the analytical solution of the SI model for:
    ## Time t 
    ## Transmissibility beta
    ## Initial condition I0
    return (I0*np.exp(beta*t))/((1 - I0) + I0*np.exp(beta*t))
```

Then we use this function to obtain the trajectory of this model over time $t$ and for a given value of $\beta$ and $I_0$. By plotting, we can observe that the proportion of infected individuals grows monotonically over time until all the population is infected. 

```python
t = np.linspace(0, 500, 500)
beta = 0.1
I0 = 0.1
y = model_SI(t, beta, I0)
```

```python
plt.plot(t, y, label = 'Infected')
plt.plot(t, 1 - y, label = 'Susceptible')
plt.legend()
plt.xlabel('Time')
plt.ylabel('Proportion of the population')
plt.show()
```

We will then separate the first 50 points of the infected curve for training an ordinary neural network, which will take the time $t$ as an input and will aim to predict the infected proportion. This will be done by finding optimal network parameters such that the loss function is minimized. The optimizer that will be used is the Adam optimizer, already available from the pytorch package. The loss function will be the Mean Squared Error (MSE), defined as:

\begin{aligned}
\mathcal{L}(\theta) = \sum_t (\hat{I}(\theta) - I)^2,
\end{aligned}

where $\theta$ are the network parameters, I is the number of infected provided to train the network and $\hat{I}$ is the network prediction. We will train the network for a total of 10.000 epochs. 
Regarding network architecture, we will use a network with 3 hidden layers of 32 neurons, using the $tanh$ as the activation function in all layers. 

```python
# defining training data
df = pd.DataFrame({'t': t[0:50], 'infected': y[0:50]}) 
t_nn = torch.tensor(df['t'], dtype = torch.float32).view(-1, 1)
i_nn = torch.tensor(df['infected'], dtype = torch.float32).view(-1, 1)

# creating network and defining optimizer
model = NNetwork(1, 1, 32, 3)
optimizer = torch.optim.Adam(model.parameters(), lr = 1e-4)

# training
n_epochs = 10000
loss_epochs = []

for epochs in range(n_epochs):
    optimizer.zero_grad()
    i_pred = model(t_nn)
    loss = torch.mean((i_pred - i_nn)**2)
    loss_epochs.append(loss.detach().cpu().numpy())
    loss.backward()
    optimizer.step()
```

After training, we plot the loss function, which seems to have a normal behaviour. 

```python
plt.plot(loss_epochs)
plt.xlabel('Epoch')
plt.ylabel('Neural Network Loss')
plt.show()
```

And then we are ready to plot the actual $I(t)$ trajectory, alongside the predictions from our ordinary neural network. As we can see, even though qualitatively the behaviour of the infected number is similar, the neural network predicts proportions of infected above the unity, which makes no physical sense.

```python
df_total = pd.DataFrame({'t': t, 'infected': y})
t_total = torch.tensor(df_total['t'], dtype = torch.float32).view(-1, 1)
i_total = torch.tensor(df_total['infected'], dtype = torch.float32).view(-1, 1)    
i_pred = model(t_total) 
df_total['i_pred'] = i_pred.detach().cpu().numpy()

plt.plot(df_total['t'], df_total['infected'], label = 'Ground Truth')
plt.plot(df_total['t'][0:50], df_total['infected'][0:50], '-o', label = 'Train Data')
plt.plot(df_total['t'], df_total['i_pred'], label = 'Neural Network')
plt.axhline(1, linestyle = '--', color = 'red')
plt.xlabel('Time')
plt.ylabel(r'$I(t)$')
plt.legend()
#plt.yscale('log')
plt.show()
```

## Physics Inspired Neural Networks (PINNs)

Retaining a similar form, PINNs are a tool for solving and inferring solutions of differential equations with neural networks. They do this by complementing the loss function we introduced before with a structure depending on the differential equation under study. The loss function of a PINN has three basic components:

\begin{aligned}
\mathcal{L} = \mathcal{L_{\mathrm{data}}} + \mathcal{L_{\mathcal{boundary}}} + \mathcal{L_{\mathrm{physics}}}
\end{aligned}

The first component, termed data loss, is the same loss introduced before, which measures the distance between training data and neural network predictions. Using, for example, MSE, we have that:

\begin{aligned}
\mathcal{L_{\mathrm{data}}} = \sum_t (\hat{I(t)} - I(t))^2
\end{aligned}

The second component, termed boundary loss, aims to enforce initial conditions such that the network is able to find the true solution, corresponding to the given initial condition. As such, this loss measures the distance between the initial condition provided to the network and the network prediction for $t = 0$. Mathematically, this can be enforced by the loss:

\begin{aligned}
\mathcal{L_{\mathrm{boundary}}} = (\hat{I}(t = 0) - I_0)^2
\end{aligned}

Finally, the third and most important term, which is often called the physics or the ODE loss, uses the automatic differentiation capability of several neural network frameworks to compute:

\begin{aligned}
\mathcal{L_{\mathrm{physics}}} = \sum_t \widehat{\frac{dI}{dt} (t)} - \beta (1 - \hat{I(t)}) \hat{I(t)},
\end{aligned}
thus enforcing the dynamics of the ODE in question upon the neural network predictions. 

In our next step, we will use the same training data, basic network architecture and optimizer in order to train a neural network with a loss of his form, thus a PINN, and compare its performance with the one of the ordinary neural network for the same task.


### Defining training data 

We will first define the necessary training data by creating tensors for training, similar to what we did for the classical neural network. First, we define the initial conditions:  

```python
t_initial = torch.tensor([0.0], dtype=torch.float32).view(-1, 1)
i_initial = torch.tensor([I0], dtype = torch.float32).view(-1, 1)
```

Secondly, we define the remaining data needed for training. We will use the first 50 points of the previously calculated trajectory. It is important to note that, when defining the time vector, we should include the <code>.requires_grad_(True)</code> method. This enables PyTorch to compute gradients with respect to this tensor, which is necessary for calculating the $\mathcal{L}_{\mathrm{ODE}}$ loss. 

```python
t_nn = torch.tensor(df['t'], dtype = torch.float32).view(-1, 1).requires_grad_(True)
i_nn = torch.tensor(df['infected'], dtype = torch.float32).view(-1, 1)
```

Finally, we define the network architecture and parameters necessary for training. In each epoch, we compute the corresponding loss, taking into account the three components necessary for PINNs. 

```python
model = NNetwork(1, 1, 32, 3)
optimizer = torch.optim.Adam(model.parameters(), lr = 1e-4)
n_epochs = 10000
loss_epochs = []

for epoch in range(n_epochs):
    
    optimizer.zero_grad()
    i_pred = model(t_nn)
    i_pred_initial = model(t_initial)
    
    # ODE Loss
    dx = torch.autograd.grad(i_pred, t_nn, torch.ones_like(i_pred), create_graph=True)[0]
    loss_ode_aux = dx - beta*i_pred*(1 - i_pred)
    loss_ode = torch.mean(loss_ode_aux**2)
    
    # Data Loss
    loss_data = torch.mean((i_pred - i_nn)**2)
    
    # Boundary condition Loss
    loss_boundary = torch.mean((i_pred_initial - i_initial)**2)
    
    # Calculate total loss
    loss = loss_ode + loss_boundary + loss_data 
    loss_epochs.append(loss.item())
    loss.backward()
    optimizer.step()
```

We can plot the loss, as done for any neural network, independent of its class.  

```python
plt.plot(loss_epochs)
plt.xlabel('Epoch')
plt.ylabel('Neural Network Loss')
plt.show()
```

As we did before, we can plot the true values of $I(t)$ over time, along with the predictions from the PINN and the ordinary neural network. Furthermore, we highlight the values used for training. As can be observed from the following plot, the PINN is able to better approximate $I(t)$, even using the same data for training, as well as a similar network architecture.  

```python
i_pred = model(t_total)
df_total['i_pred_pinn'] = i_pred.detach().cpu().numpy()

plt.plot(df_total['t'], df_total['infected'], label = 'Ground Truth')
plt.plot(df_total['t'][0:50], df_total['infected'][0:50], '-o', label = 'Training Data')
plt.plot(df_total['t'], df_total['i_pred'], label = 'NN')
plt.plot(df_total['t'], df_total['i_pred_pinn'], label = 'PINN')
plt.xlabel('Time')
plt.ylabel(r'$I(t))$')
plt.legend()
#plt.yscale('log')
plt.show()
```
